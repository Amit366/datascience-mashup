# Installing Required Libraries

Let's see which libraries are required to plot graphs in Seaborn. Make sure you install following libraries to proceed further.
1. Pandas

3. Seaborn

```python
#IMPORTING LIBRARIES
import pandas as pd #for loading datasets
from matplotlib import pyplot as plt #for customisation
%matplotlib inline
# Seaborn for plotting and styling
import seaborn as sns
```

# 1. Implot

‘lmplot’ is the most basic plot which shows a line along a 2-dimensional plane and is used to see the correlation between two attributes plotted against the x-axis and y-axis. Here we are considering two variables Applicant income and Loan Amount. 

Seaborn also allows you to set the height, colour palette, etc. properties for the plot generated.

```python
#importing the dataset
df=pd.read_csv('Loan Prediction Dataset.csv')
df.head
```

```python
#WITHOUT FITTING LINEAR REGRESSION
sns.lmplot(x='ApplicantIncome', y='LoanAmount', data=df,
           fit_reg=False, # No regression line
          )
```

```python
#WITH FITTING LINEAR REGRESSION
sns.lmplot(x='ApplicantIncome', y='LoanAmount', data=df,
           fit_reg=True, #  regression line
          )
```

# 2. Barplot

A barplot is basically used to aggregate the categorical data according to some methods and by default it’s the mean. It can also be understood as a visualization of the group by action. To use this plot we choose a categorical column for the x-axis and a numerical column for the y-axis, and we see that it creates a plot taking a mean per categorical column.

```python
#Plotting Applicant income and loan status to see if applicant income affects his loan status
sns.barplot(x ='Loan_Status', y ='ApplicantIncome', data = df)
```

# 3. Count Plot

A countplot basically counts the categories and returns a count of their occurrences. It is one of the simplest plots provided by the seaborn library

```python
#Plotting Loan status to count people with acceptance and rejection of Loan
sns.countplot(x ='Loan_Status', data = df) #Y= yes, N=no
```

# 4. Box Plot

Box Plot is the visual representation of the depicting groups of numerical data through their quartiles. Boxplot is also used to detect the outlier in the data set.

```python
sns.boxplot(x='Property_Area', y='ApplicantIncome', data=df)
```

# 5. KDE Plot

A Kernel Density Estimate plot is used to visualize the Probability density distribution of univariate data. In simple terms, we can use it to know the spread/distribution of the data.

```python
sns.kdeplot(df['ApplicantIncome'])
```

# 6.Dist Plot

Distplot is the most convenient way of visualizing the distribution of the dataset and the skewness of the data. It is a combination of kdeplot and histograms.

```python
sns.distplot(df['ApplicantIncome'])
```

# 7. HeatMap

Heatmaps are used to display the correlations of different numerical attributes in a dataset. In heatmaps, colour scheme plays an important role in visualizing whether the relationship is positive or negative. 
For creating a heatmap we will create a Correlation matrix and pass it to the heatmap parameter. We will also set the annotation to true so the value of the relationship is also visible.

```python
sns.heatmap(df.corr(), annot=True)
```

These are some of the basic plots which we can visualize using Seaborn, and are helpful in data analysis. Seaborn has the advantage of manipulating the graphs and plots by applying different parameters. 
Some of the important parameters are:

1. set_style: It is used to set the aesthetics style of the plots, mainly affects the properties of the grid and axes.
2. hue: It is used for deciding which column of the dataset will be used for colour encoding. 
3. palette:  It is used to select the colour palette which we want to use for our visualization. Seaborn has six pre-defined colour palettes namely: “pastel”, “muted”, “bright”, “deep”, “colorblind” and “dark”.
4. height: It sets the height of the visualization figure.

Other than these properties all the graphs have some of their internal properties which can be altered accordingly.
