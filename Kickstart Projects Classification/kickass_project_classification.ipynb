```python
y = df['state']
y.shape
```

```python
df = df.drop('state', 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df,y, test_size = 0.1,random_state=42)

print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from lightgbm import LGBMClassifier
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
```

```python
print('Accuracy for Logistic Regression: {}'.format(round(logreg.score(X_test, Y_test) * 100,3)))
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
```

```python
ans = round(knn.score(X_test, Y_test) * 100,3)
```

```python
print('Accuracy for K-Nearest Neighbors Classifier : {}'.format(ans))
```

```python
abc = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1))
abc.fit(X_train, Y_train)
```

```python
ans = round(abc.score(X_test, Y_test) * 100,3)
```

```python
print('Accuracy for AdaBoostClassifier : {}'.format(ans))
```

```python
gbc = GradientBoostingClassifier(max_depth=1, random_state=0)
gbc.fit(X_train, Y_train)
```

```python
ans = round(gbc.score(X_test, Y_test) * 100,3)
```

```python
print('Accuracy for GradientBoostingClassifier : {}'.format(ans))
```

```python
bagging = BaggingClassifier(KNeighborsClassifier(n_neighbors=8))
bag = bagging.fit(X_train,Y_train)
```

```python
ans = round(bag.score(X_test, Y_test) * 100,3)
```

```python
print('Accuracy for BaggingClassifier : {}'.format(ans))
```
